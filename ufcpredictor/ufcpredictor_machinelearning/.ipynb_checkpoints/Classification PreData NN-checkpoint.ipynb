```python
import numpy as np
import pandas as pd

df = pd.read_csv('archive/preprocessed_data.csv')

df = df.dropna()
df.isnull().sum()

dfBlue = df.loc[df['Winner'] == 'Blue']

df = pd.concat([df,dfBlue]).sample(frac=1)
```

```python
df['Winner'].value_counts()
```

```python
df['title_fight'] = [1 if tb == 'True' else 0 for tb in df['title_bout']]
df['isRedWin'] = [1 if Winner == 'Red' else 0 for Winner in df['Winner']]
df['B_win_by_KO/TKO'] = df[['B_win_by_KO/TKO','B_win_by_TKO_Doctor_Stoppage']].sum(axis=1)
df['R_win_by_KO/TKO'] = df[['R_win_by_KO/TKO','R_win_by_TKO_Doctor_Stoppage']].sum(axis=1)
```

```python
df.drop('Winner', axis=1, inplace=True)
df.drop('title_bout', axis=1, inplace=True)
df.drop('B_win_by_TKO_Doctor_Stoppage', axis=1, inplace=True)
df.drop('B_avg_opp_KD', axis=1, inplace=True)
df.drop('B_avg_opp_SIG_STR_pct', axis=1, inplace=True)
df.drop('B_avg_opp_TD_pct', axis=1, inplace=True)
df.drop('B_avg_opp_SUB_ATT', axis=1, inplace=True)
df.drop('B_avg_opp_REV', axis=1, inplace=True)
df.drop('B_avg_opp_SIG_STR_att', axis=1, inplace=True)
df.drop('B_avg_opp_SIG_STR_landed', axis=1, inplace=True)
df.drop('B_avg_opp_TOTAL_STR_att', axis=1, inplace=True)
df.drop('B_avg_opp_TOTAL_STR_landed', axis=1, inplace=True)
df.drop('B_avg_opp_TD_att', axis=1, inplace=True)
df.drop('B_avg_opp_TD_landed', axis=1, inplace=True)
df.drop('B_avg_opp_HEAD_att', axis=1, inplace=True)
df.drop('B_avg_opp_HEAD_landed', axis=1, inplace=True)
df.drop('B_avg_opp_BODY_att', axis=1, inplace=True)
df.drop('B_avg_opp_BODY_landed', axis=1, inplace=True)
df.drop('B_avg_opp_LEG_att', axis=1, inplace=True)
df.drop('B_avg_opp_LEG_landed', axis=1, inplace=True)
df.drop('B_avg_opp_DISTANCE_att', axis=1, inplace=True)
df.drop('B_avg_opp_DISTANCE_landed', axis=1, inplace=True)
df.drop('B_avg_opp_CLINCH_att', axis=1, inplace=True)
df.drop('B_avg_opp_CLINCH_landed', axis=1, inplace=True)
df.drop('B_avg_opp_GROUND_att', axis=1, inplace=True)
df.drop('B_avg_opp_GROUND_landed', axis=1, inplace=True)
df.drop('B_avg_opp_CTRL_time(seconds)', axis=1, inplace=True)
df.drop('B_total_time_fought(seconds)', axis=1, inplace=True)
df.drop('B_total_rounds_fought', axis=1, inplace=True)
df.drop('B_total_title_bouts', axis=1, inplace=True)
df.drop('B_current_win_streak', axis=1, inplace=True)
df.drop('B_current_lose_streak', axis=1, inplace=True)
df.drop('B_longest_win_streak', axis=1, inplace=True)
df.drop('B_draw', axis=1, inplace=True)
df.drop('B_Stance_Open Stance', axis=1, inplace=True)
df.drop('B_Stance_Orthodox', axis=1, inplace=True)
df.drop('B_Stance_Sideways', axis=1, inplace=True)
df.drop('B_Stance_Southpaw', axis=1, inplace=True)
df.drop('B_Stance_Switch', axis=1, inplace=True)
df.drop('B_avg_REV', axis=1, inplace=True)

df.drop('R_win_by_TKO_Doctor_Stoppage', axis=1, inplace=True)
df.drop('R_avg_opp_KD', axis=1, inplace=True)
df.drop('R_avg_opp_SIG_STR_pct', axis=1, inplace=True)
df.drop('R_avg_opp_TD_pct', axis=1, inplace=True)
df.drop('R_avg_opp_SUB_ATT', axis=1, inplace=True)
df.drop('R_avg_opp_REV', axis=1, inplace=True)
df.drop('R_avg_opp_SIG_STR_att', axis=1, inplace=True)
df.drop('R_avg_opp_SIG_STR_landed', axis=1, inplace=True)
df.drop('R_avg_opp_TOTAL_STR_att', axis=1, inplace=True)
df.drop('R_avg_opp_TOTAL_STR_landed', axis=1, inplace=True)
df.drop('R_avg_opp_TD_att', axis=1, inplace=True)
df.drop('R_avg_opp_TD_landed', axis=1, inplace=True)
df.drop('R_avg_opp_HEAD_att', axis=1, inplace=True)
df.drop('R_avg_opp_HEAD_landed', axis=1, inplace=True)
df.drop('R_avg_opp_BODY_att', axis=1, inplace=True)
df.drop('R_avg_opp_BODY_landed', axis=1, inplace=True)
df.drop('R_avg_opp_LEG_att', axis=1, inplace=True)
df.drop('R_avg_opp_LEG_landed', axis=1, inplace=True)
df.drop('R_avg_opp_DISTANCE_att', axis=1, inplace=True)
df.drop('R_avg_opp_DISTANCE_landed', axis=1, inplace=True)
df.drop('R_avg_opp_CLINCH_att', axis=1, inplace=True)
df.drop('R_avg_opp_CLINCH_landed', axis=1, inplace=True)
df.drop('R_avg_opp_GROUND_att', axis=1, inplace=True)
df.drop('R_avg_opp_GROUND_landed', axis=1, inplace=True)
df.drop('R_avg_opp_CTRL_time(seconds)', axis=1, inplace=True)
df.drop('R_total_time_fought(seconds)', axis=1, inplace=True)
df.drop('R_total_rounds_fought', axis=1, inplace=True)
df.drop('R_total_title_bouts', axis=1, inplace=True)
df.drop('R_current_win_streak', axis=1, inplace=True)
df.drop('R_current_lose_streak', axis=1, inplace=True)
df.drop('R_longest_win_streak', axis=1, inplace=True)
df.drop('R_draw', axis=1, inplace=True)
df.drop('R_Stance_Open Stance', axis=1, inplace=True)
df.drop('R_Stance_Orthodox', axis=1, inplace=True)
df.drop('R_Stance_Sideways', axis=1, inplace=True)
df.drop('R_Stance_Southpaw', axis=1, inplace=True)
df.drop('R_Stance_Switch', axis=1, inplace=True)
df.drop('R_avg_REV', axis=1, inplace=True)
```

```python
df = df.astype(float)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('isRedWin',axis=1)
y = df['isRedWin']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
X_train.shape, X_test.shape
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled[:3]
```

```python
import tensorflow as tf
tf.random.set_seed(30)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])
```

```python
model.compile(
    loss = tf.keras.losses.binary_crossentropy,
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.03),
    metrics=[
        tf.keras.metrics.BinaryAccuracy(name='accuracy'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)
```

```python
history = model.fit(X_train_scaled, y_train, epochs = 10)
```

```python
pred = model.predict(X_test_scaled)
pred
```

```python
pred_classes = [1 if val > 0.5 else 0 for val in np.ravel(pred)]
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, pred_classes))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

print(f'Accuracy: {accuracy_score(y_test,pred_classes):.2f}')
print(f'Precision: {precision_score(y_test,pred_classes):.2f}')
print(f'Recall: {recall_score(y_test,pred_classes):.2f}')
```

```python
from sklearn.metrics import roc_curve, auc, RocCurveDisplay

fpr, tpr, threshold = roc_curve(y_test, pred)
roc_auc = auc(fpr, tpr)
display = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,
                                   estimator_name='example estimator')
display.plot()
```
